```python
from google.colab import auth
auth.authenticate_user()
import gspread
from oauth2client.client import GoogleCredentials
gc = gspread.authorize(GoogleCredentials.get_application_default())
```

```python
wb = gc.open_by_url('https://docs.google.com/spreadsheets/d/17VjM1bkrjSVa-UbC-1OJ3wQS-yOI5IQL3w6mDJ5auh0/edit?usp=sharing')
sheet = wb.worksheet('bauxite_records')
data = sheet.get_all_values()
```

```python
import pandas as pd
df = pd.DataFrame(data)
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header #reshaping the column names
```

```python
df.head()
```

```python
df = df.iloc[17:] #data from 1977 onwards
for items in df.columns[1:]: 
  df[items] = df[items].astype(int)
df.head()
```

```python
df.set_index("Year",inplace=True)
```
